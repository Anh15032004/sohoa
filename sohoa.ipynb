```python
import pandas as pd
import numpy as np

# Dữ liệu giả định (lấy từ bảng của bạn)
data = {
    "Hours_Worked": [40, 45, 38, 42, None, 39, 41, 44, 37, 40,
                     46, 38, 43, None, 39, 41, 45, 37, None, 44],
}

# Tạo DataFrame
df = pd.DataFrame(data)

# Điền giá trị thiếu bằng trung bình (tạm thời để tính toán)
df["Hours_Worked"] = df["Hours_Worked"].fillna(df["Hours_Worked"].mean())

# Tính độ lệch chuẩn và trung vị
std_hours_worked = np.std(df["Hours_Worked"])
median_hours_worked = np.median(df["Hours_Worked"])

print("Độ lệch chuẩn của Hours_Worked:", std_hours_worked)
print("Trung vị của Hours_Worked:", median_hours_worked)
```

```python
# Tính IQR
Q1 = df["Performance"].quantile(0.25)  # Quartile 1
Q3 = df["Performance"].quantile(0.75)  # Quartile 3
IQR = Q3 - Q1  # Interquartile Range

# Ngưỡng dưới và trên để xác định outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Lọc ra các nhân viên có Performance nằm ngoài ngưỡng
outliers = df[(df["Performance"] < lower_bound) | (df["Performance"] > upper_bound)]
print("Những nhân viên outliers:")
print(outliers)
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Dữ liệu mẫu
data = {
    "Employee_ID": ["E001", "E002", "E003", "E004", "E005", "E006", "E007", "E008", "E009", "E010", 
                    "E011", "E012", "E013", "E014", "E015", "E016", "E017", "E018", "E019", "E020"],
    "Performance": [88, None, 75, 92, 68, 85, None, 79, 90, 73,
                    86, None, 95, 70, 82, 78, 89, 71, 87, 93],
    "Hours_Worked": [40, 45, 38, 42, None, 39, 41, 44, 37, 40,
                     46, 38, 43, None, 39, 41, 45, 37, None, 44],
    "Training_Hours": [5, 3, None, 6, 4, 5, 3, 4, None, 5,
                       2, 6, 7, 3, 5, 4, None, 3, 5, 6]
}

df = pd.DataFrame(data)

# Điền giá trị thiếu trong Hours_Worked và Training_Hours bằng trung bình
df["Hours_Worked"] = df["Hours_Worked"].fillna(df["Hours_Worked"].mean())
df["Training_Hours"] = df["Training_Hours"].fillna(df["Training_Hours"].mean())

# Tách dữ liệu không thiếu vào tập huấn luyện
train_data = df[df["Performance"].notna()]
test_data = df[df["Performance"].isna()]

X_train = train_data[["Hours_Worked", "Training_Hours"]]
y_train = train_data["Performance"]

# Huấn luyện mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train, y_train)

# Dự đoán giá trị thiếu
X_test = test_data[["Hours_Worked", "Training_Hours"]]

# TẠO DataFrame có tên cột chính xác cho X_test
X_test_df = pd.DataFrame(X_test, columns=["Hours_Worked", "Training_Hours"])

# Sử dụng X_test_df để dự đoán
df.loc[df["Performance"].isna(), "Performance"] = model.predict(X_test_df)

print("Dữ liệu sau khi điền bằng hồi quy tuyến tính:")
print(df)
```

```python
# Điền giá trị thiếu bằng trung bình
df["Performance_Mean"] = df["Performance"].fillna(df["Performance"].mean())

print("Dữ liệu sau khi điền bằng trung bình:")
print(df)
```

Sử dụng hồi quy tuyến tính khi Hours_Worked và Training_Hours có giá trị đầy đủ.

Sử dụng trung bình khi Hours_Worked hoặc Training_Hours bị thiếu.

Xử lý linh hoạt cho dữ liệu thiếu:

Trường hợp thiếu quá nhiều dữ liệu (vd: cả Hours_Worked và Training_Hours đều thiếu), phương pháp trung bình đóng vai trò dự phòng.

```python
# Hàm điền giá trị thiếu bằng cách tiếp cận lai
def hybrid_fill(row, model, mean_performance):
    if pd.notna(row["Hours_Worked"]) and pd.notna(row["Training_Hours"]):
        return model.predict([[row["Hours_Worked"], row["Training_Hours"]]])[0]
    else:
        return mean_performance

mean_performance = df["Performance"].mean()
df["Performance_Hybrid"] = df.apply(lambda row: hybrid_fill(row, model, mean_performance), axis=1)
print("Dữ liệu sau khi điền bằng cách tiếp cận lai:")
print(df)
```

```python
import pandas as pd
import numpy as np

# Dữ liệu ban đầu
data = {
    "Employee_ID": ["E001", "E002", "E003", "E004", "E005", "E006", "E007", "E008", "E009", "E010", 
                    "E011", "E012", "E013", "E014", "E015", "E016", "E017", "E018", "E019", "E020"],
    "Performance": [88, None, 75, 92, 68, 85, None, 79, 90, 73, 86, None, 95, 70, 82, 78, 89, 71, 87, 93]
}

# Tạo DataFrame và xử lý giá trị None
df = pd.DataFrame(data)
df["Performance"] = df["Performance"].fillna(df["Performance"].mean())

# Hàm xác định nhân viên ngoài 2 độ lệch chuẩn
def detect_outliers(dataframe, column):
    mean = dataframe[column].mean()
    std_dev = dataframe[column].std()
    lower_bound = mean - 2 * std_dev
    upper_bound = mean + 2 * std_dev
    outliers = dataframe[(dataframe[column] < lower_bound) | (dataframe[column] > upper_bound)]
    return outliers

# Xác định nhân viên ngoài 2 độ lệch chuẩn
outliers = detect_outliers(df, "Performance")
print("Nhân viên có Performance ngoài 2 độ lệch chuẩn:")
print(outliers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets, transforms

# Tải dữ liệu MNIST
transform = transforms.Compose([transforms.ToTensor()])
train_dataset = datasets.MNIST(root='./data', train=True, download=True, transform=transform)

# Tính giá trị pixel trung bình cho từng chữ số
average_digits = {i: np.zeros((28, 28)) for i in range(10)}
counts = {i: 0 for i in range(10)}

for image, label in train_dataset:
    average_digits[label] += image.squeeze().numpy()
    counts[label] += 1

for digit in average_digits:
    average_digits[digit] /= counts[digit]

# Vẽ heatmap
for digit, avg_image in average_digits.items():
    plt.imshow(avg_image, cmap='hot')
    plt.title(f'Heatmap for Digit {digit}')
    plt.colorbar()
    plt.show()
```

Chiến lược đề xuất:
Batch size nhỏ (ví dụ: 32):

Batch size lớn (ví dụ: 128 hoặc 256):

Sử dụng khi có GPU mạnh mẽ (ví dụ: NVIDIA RTX 3090 hoặc tương đương).

Giảm thời gian huấn luyện tổng thể bằng cách tính toán nhanh hơn trong mỗi epoch.
Lựa chọn phù hợp
Nếu tài nguyên máy tính của bạn giới hạn (GPU dung lượng thấp), nên chọn batch size nhỏ (16 hoặc 32).

Nếu bạn có tài nguyên mạnh mẽ, hãy tận dụng batch size lớn (128 hoặc 256) để tăng tốc độ huấn luyện.

Câu 18:

```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

# Kiến trúc ANN
class MNIST_ANN(nn.Module):
    def __init__(self):
        super(MNIST_ANN, self).__init__()
        self.layers = nn.Sequential(
            nn.Linear(784, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 10)
        )  

    def forward(self, x):
        return self.layers(x)

# Chuẩn bị dữ liệu
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
train_dataset = datasets.MNIST('./data', train=True, download=True, transform=transform)
test_dataset = datasets.MNIST('./data', train=False, transform=transform)

# Tạo DataLoader
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000)

# Khởi tạo mô hình
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MNIST_ANN().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Huấn luyện với Early Stopping
train_losses, test_losses = [], []
best_loss = float('inf')
patience_counter = 0

for epoch in range(20):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images = images.view(-1, 784).to(device)
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels.to(device))
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    
    # Đánh giá trên tập test
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for images, labels in test_loader:
            images = images.view(-1, 784).to(device)
            outputs = model(images)
            test_loss += criterion(outputs, labels.to(device)).item()
    
    # Ghi nhận loss
    epoch_train_loss = running_loss/len(train_loader)
    epoch_test_loss = test_loss/len(test_loader)
    train_losses.append(epoch_train_loss)
    test_losses.append(epoch_test_loss)
    
    # Early Stopping
    if epoch_test_loss < best_loss:
        best_loss = epoch_test_loss
        patience_counter = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        patience_counter += 1
        if patience_counter >= 5:
            print(f"Early stopping at epoch {epoch+1}")
            break
    
    print(f"Epoch {epoch+1}: Train Loss: {epoch_train_loss:.4f}, Test Loss: {epoch_test_loss:.4f}")

# Vẽ biểu đồ loss
plt.plot(train_losses, label='Training Loss')
plt.plot(test_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')
plt.show()
```

Đề xuất một kiến trúc ANN khác để cải thiện hiệu suất:
Thay đổi số nơ-ron ở các tầng: Tăng số nơ-ron ở các tầng ẩn (ví dụ: tầng ẩn 1 từ 256 lên 512, tầng ẩn 2 từ 128 lên 256).

Thêm tầng ẩn: Thêm một tầng ẩn khác (ví dụ: tầng ẩn 3 với 128 nơ-ron) để tăng độ phức tạp của mô hình, giúp học được các đặc trưng phức tạp hơn.

Ví dụ về kiến trúc thay đổi:

```python
class ImprovedANN(nn.Module):
    def __init__(self):
        super(ImprovedANN, self).__init__()
        self.fc1 = nn.Linear(784, 512)
        self.fc2 = nn.Linear(512, 256)
        self.fc3 = nn.Linear(256, 128)
        self.fc4 = nn.Linear(128, 10)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = x.view(-1, 28 * 28)
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.relu(self.fc3(x))
        x = self.fc4(x)
        return x
```

Lý do:
Tăng số nơ-ron hoặc số tầng giúp mô hình có khả năng học các đặc trưng phức tạp hơn trong dữ liệu.

Tuy nhiên, điều này cũng làm tăng nguy cơ overfitting. Do đó, bạn có thể áp dụng các kỹ thuật như Dropout hoặc Batch Normalization để giảm overfitting.
